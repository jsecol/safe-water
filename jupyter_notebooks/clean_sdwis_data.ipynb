```python
# import sys
import os
import pandas as pd
import time
import csv

pd.options.display.max_columns = None


#### cool from data-exploration.ipynb
# display(df.sample(10))
# display(df.nunique().sort_values())
```

```python
DATA_PATH1 = 'C:/Users/Jim/Files/CFB/data/swdis/WATER_SYSTEM.csv' 

start_time = time.time()

DATA1 = pd.read_csv(DATA_PATH1, header=0, sep=',', error_bad_lines=False,
                         dtype={0:'str'}, quotechar='"', low_memory=False, index_col=False)


# with open('C:/Users/Jim/Files/CFB/data/swdis/bad_lines.txt', 'w') as fp:
#     sys.stderr = fp
#     pd.read_csv('DATA_PATH1', sep=',', error_bad_lines=False, 
#                         dtype={0:'str'}, quotechar='"', low_memory=False, index_col=False)


print(time.time() - start_time)

#compare length of rows by length of columns in df
lencols1 = len(DATA1.columns) + 1 #csv.reader counts 1 less for some reason
print(lencols1)

result1 = []
with open(DATA_PATH1) as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    for row in reader:
        if (len(row) > lencols1):
            print("Length of row is: %r" % len(row) )
            print(row)
            print(reader.line_num)
            result1.append(str(reader.line_num))


```

```python
print(result1) # maybe can read in just these lines as separate strings into a custom df
```

```python
print("rows lost =",len(result1), "out of", len(DATA1.index), 
      "(", 100*round(len(result1)/(len(DATA1.index)+len(result1)), 10),"%",")")
```

```python
DATA1.head(5)
```

```python
DATA1 = DATA1.replace('"', '', regex=True)
```

```python
# DATA1.str.replace('[^a-zA-Z0-9]', '') #too drastic, amybe
```

```python
fpath1 = 'C:/Users/Jim/Files/CFB/data/swdis/WATER_SYSTEM_clean.csv'

DATA1.to_csv(fpath1, sep="|", index=False, header=1, encoding='utf-8', quoting=csv.QUOTE_MINIMAL)
```
