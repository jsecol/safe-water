```python
import os
import pandas as pd
import time
import csv

pd.options.display.max_columns = None
pd.options.display.max_rows = 200
```

```python
ECHODATA_PATH = '../data/echo_exporter.zip'

start_time = time.time()

ECHOdata1 = pd.read_csv(ECHODATA_PATH, compression='zip', header=0, sep=',', 
                        dtype={0:'str'}, quotechar='"', low_memory=False, index_col=False) #, nrows=5

print(time.time() - start_time)

ECHOdata1.shape
```

```python
start_time = time.time()

ECHOdata1 = ECHOdata1.replace('\"', '\'', regex=True)

print(time.time() - start_time)

ECHOdata1.shape
```

Now to save with "|" as delimiter, etc.

```python
start_time = time.time()

fpath1 = 'C:/Users/Jim/Files/CFB/echo_exporter/ECHO_EXPORTER_clean.csv'

ECHOdata1.to_csv(fpath1, sep="|", index=False, header=1, encoding='utf-8', quoting=csv.QUOTE_MINIMAL)

print(time.time() - start_time)
```
