```python
import os
import pandas as pd

DATABASE_PATH = '../data/simple_time_based_model.zip'

# Allow us to display all the columns in a dataframe
pd.options.display.max_columns = None
```

```python
df = pd.read_csv(DATABASE_PATH, compression='zip', header=0, sep=',', quotechar='"', low_memory=False, index_col=False)
df.shape
```

```python
df.head(5)
```

```python
df[df.PWSID == '010106001']
```

```python
df_id_year = df.loc[:,['PWSID', 'YEAR']]
```

```python
df_id_year.head()
```

```python
df_id_year = pd.get_dummies(df_id_year, columns=['YEAR'])
```

```python
df_id_year[df_id_year.PWSID == '010106001']
```

```python
# Note: tried grouping on PWS_TYPE_CODE too, and got a different # of rows 
# (small diff, but maybe PWS_TYPE_CODE can change for a PWSID? Id which one(s) later)

df_id_year2 = df_id_year.groupby(['PWSID'], 
                              as_index=False)[['YEAR_2014.0', 'YEAR_2015.0', 
                                               'YEAR_2016.0', 'YEAR_2017.0', 'YEAR_2018.0']] \
                                .sum().rename(columns={'YEAR_2014.0':'YEAR_2014', 
                                                       'YEAR_2015.0':'YEAR_2015',
                                                       'YEAR_2016.0':'YEAR_2016',
                                                       'YEAR_2017.0':'YEAR_2017',
                                                       'YEAR_2018.0':'YEAR_2018'})
```

```python
df_id_year2.describe()
```

### I. Matching up the data
1. SDWA_IDS can be a mess, with multiple ID's some with two-letter state codes to start, others not
2. Starting with filtering down ECHO data to those that don't match up with simple_time_based_model.csv PWSID
