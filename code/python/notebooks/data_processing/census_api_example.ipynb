```python
state_codes = {'CT': '09', 'MA': '25', 'ME': '23', 'NH': '33', 'RI': '44', 'VT': '50'}
```

```python
states = list(state_codes.values())
```

```python
states
```

```python
cendatlst = list()
for s in states:
    s_filter = 'state: %s' % (s)
    cendat = c.acs5.get(('NAME', 'B01001_001E'), 
                        {'for': 'county subdivision:*', 'in': s_filter}, year = 2017)
    cendatlst.append(cendat)
    
```
